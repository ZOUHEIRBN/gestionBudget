```python
import json

with open('../database.json', 'r', encoding='utf8') as f:
    d = json.load(f)
    
d
```

```python
import random
def generate_sequence(n=64):
    s = ""
    for i in range(n):
        s = s + chr(random.randrange(65, 65+26))
    return s

generate_sequence()
```

```python
for m in d['markets']:
    m['id'] = generate_sequence()
```
